```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python

def rand_dot():
    arr1 = np.random.rand(1000)
    arr2 = np.random.rand(1000)
    return np.dot(arr1, arr2)

dots = [rand_dot() for _ in range(1000)]
dots_cum = np.cumsum(dots)
dots_avg = [(dot_cum/i)/250 for i, dot_cum in enumerate(dots_cum, start=1)]
skip = int(len(dots_avg) / 10)
print(dots_avg[::skip])
plt.plot(dots_avg)
```
