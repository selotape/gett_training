```python
from matplotlib import cm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df_taxi =pd.read_csv(r'./yellow_tripdata_2016-01_small.zip')
```

```python
df_taxi.dtypes
```

```python
#draw as scatter plot of the pickup locations (latitude is x, longitude is y)
plt.figure(figsize=(10,10))
plt.scatter(df_taxi['pickup_latitude'], df_taxi['pickup_longitude'], s=0.01)
plt.show()
```

```python
#draw as scatter plot of the dropoff locations (latitude is x, longitude is y)
plt.figure(figsize=(20, 20))
size_args = {'s':0.1, 'lw':0}
plt.scatter(df_taxi['pickup_latitude'], df_taxi['pickup_longitude'], color='green', **size_args)
plt.scatter(df_taxi['dropoff_latitude'], df_taxi['dropoff_longitude'], color='red', **size_args)
plt.show()
```

## Histograms 

```python
df_taxi['tpep_pickup_datetime'] = df_taxi['tpep_pickup_datetime'].apply(pd.Timestamp)
df_taxi['hour_of_day'] = df_taxi['tpep_pickup_datetime'].dt.hour
```

```python
#plot a histogram of rides by hour (it should have 24 bins)
plt.hist(df_taxi['hour_of_day'], bins=24)
plt.show()
```

```python
#plot a histogram of ride prices, consider a reasonable range of prices
plt.hist(df_taxi['total_amount'], bins=range(0, 40), rwidth=0.8 )
plt.show()
```
